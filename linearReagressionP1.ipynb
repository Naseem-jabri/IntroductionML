```python
#required libraries
import numpy as np #math fun + arrays
import pandas as pd #data (writing and struture)
import matplotlib.pyplot as plt #vesulazation
```

```python
# read file
#use csv beacuse file is csv
data= pd.read_csv('/content/Salary_Data.csv')
```

```python
#return data
data
```

```python
#x is feature and y is lable
#تعيين قيم للمتغيرات

x= data['YearsExperience'].values
y= data ['Salary'].values

#returen all data in the row and put the varible
# x and y is arrays
print(x)
print(y)

#number is rows
#Two dimension is the number of rows and columns and has an index of zero=rows and one=column
print(x.shape)
print(y.shape)
```

```python
#

plt.figure(figsize=(8,6))
plt.title('data distribution')
plt.scatter(x,y)
plt.xlabel('YearsExperience') # Changed plt.x.lable to plt.xlabel
plt.ylabel('Salary') # Changed plt.y.lable to plt.ylabel
plt.show()

#The result will be a linear regression
```

```python
L_rate= 0.0001 #choose small value
iterations= 1000 #choose big value

#initialization (افتراض قيم ابتدائيه للثيتا)
theta_1=0
theta_0=0

#number of samples in the dataset
n= x.shape[0]

#an empty list to store the error  in each iteration
losses=[]

#The beginning of the learning process

for i in range(iterations):
  h_x= theta_0 + theta_1*x #linear equation = h(x)= theta0 + theta1 and X is an array containing YearsExperience values
  #the result will be array

  # Keeping track of the error decrease
  mse= (1/n)*np.sum((h_x -y)**2)    #n=30 and use fun sum in np
  losses.append(mse)

  #derivatives
  d_theta0= (2/n)*np.sum(h_x-y)
  d_theta1= (2/n)*np.sum(x*(h_x-y))

  #update value

theta_1= theta_1 - L_rate*d_theta1
theta_0= theta_0 - L_rate*d_theta0




```

```python
print("theta_0", theta_0)
print("theta_1", theta_1)
print("MSE", mse)
```

```python
new_x= 5
prediction_model = theta_0 + theta_1*new_x
print("score", prediction_model)
```

```python
plt.title('loss value')
plt.plot(losses)
plt.xlabel('iteraction')
plt.ylabel('loss')
print('initial', losses[0])
print('final', losses[-1])
```
